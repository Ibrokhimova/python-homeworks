```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-10, 10, 400)
y = x**2 - 4*x + 4
plt.plot(x, y, color='blue', label='f(x) = x² - 4x + 4')
plt.title('')
plt.xlabel('')
plt.ylabel('')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 2 * np.pi, 400)
sin_x = np.sin(x)
cos_x = np.cos(x)
plt.plot(x, sin_x, color='green', linestyle='-', marker='s', label='sin(x)')
plt.plot(x, cos_x, color='orange', linestyle=':', marker='^', label='cos(x)')
plt.title('Sin(x) and cos(x) functions')
plt.xlabel('x (radian)')
plt.ylabel('Function valus')
plt.legend()
plt.grid(True)
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x_cubic = np.linspace(-10, 10, 400)
x_sin = np.linspace(-2 * np.pi, 2 * np.pi, 400)
x_exp = np.linspace(-2, 2, 400)
x_log = np.linspace(0, 10, 400)
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
axs[0, 0].plot(x_cubic, x_cubic**3, color='brown')
axs[0, 0].set_title('f(x) = x³')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('f(x)')
axs[0, 0].grid(True)
axs[0, 1].plot(x_sin, np.sin(x_sin), color='darkgreen')
axs[0, 1].set_title('f(x) = sin(x)')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('f(x)')
axs[0, 1].grid(True)
axs[1, 0].plot(x_exp, np.exp(x_exp), color='navy')
axs[1, 0].set_title('f(x) = eˣ')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('f(x)')
axs[1, 0].grid(True)
axs[1, 1].plot(x_log, np.log(x_log + 1), color='darkorange')
axs[1, 1].set_title('f(x) = log(x + 1)')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('f(x)')
axs[1, 1].grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.uniform(0, 10, 100)
y = np.random.uniform(0, 10, 100)
colors = np.random.rand(100)
plt.figure(figsize=(8, 6))
plt.scatter(x, y, c=colors, marker='o', edgecolors='black')
plt.title('Random 2D Scatter Plot')
plt.xlabel('x value')
plt.ylabel('y value')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
data = np.random.normal(loc=0, scale=1, size=1000)
plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, color='dodgerblue', edgecolor='black', alpha=0.7)
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.cos(X**2 + Y**2)
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

grafik = ax.plot_surface(X, Y, Z, cmap='coolwarm', edgecolor='none')

ax.set_title('3D : $f(x, y) = \cos(x^2 + y^2)$')
ax.set_xlabel('X ')
ax.set_ylabel('Y ')
ax.set_zlabel('Z ')
fig.colorbar(grafik, shrink=0.5, aspect=10)

plt.show()
```

```python
import matplotlib.pyplot as plt
mahsulotlar = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
savdo = [200, 150, 250, 175, 225]
ranglar = ['deepskyblue', 'tomato', 'mediumseagreen', 'orange', 'orchid']
plt.figure(figsize=(9, 6))
plt.bar(mahsulotlar, savdo, color=ranglar)
plt.title('Sales results by prodect')
plt.xlabel('Products')
plt.ylabel('Number of sales')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

time_periods = ['T1', 'T2', 'T3', 'T4']
category_A = [30, 40, 35, 50]
category_B = [20, 25, 30, 35]
category_C = [10, 15, 20, 25]
x = np.arange(len(time_periods))
plt.figure(figsize=(9, 6))
plt.bar(x, category_A, label='Category A', color='skyblue')
plt.bar(x, category_B, bottom=category_A, label='Category B', color='lightgreen')
plt.bar(x, category_C, bottom=np.array(category_A) + np.array(category_B), label='Category C', color='salmon')

plt.xticks(x, time_periods)
plt.xlabel('Time Periods')
plt.ylabel('Values')
plt.title('Category Contribution Over Time')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```
