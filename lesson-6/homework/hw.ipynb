Zero Check Decorator

```python
def check(func):
    def wrapper(a, b):
        if b == 0:
            return "Denominator can't be zero"
        return func(a, b)
    return wrapper

@check
def div(a, b):
    return a / b

print(div(6, 2)) 
print(div(6, 0))  
```

Employee Records Manager


```python
def add_employee():
    f = open("employees.txt", "a")
    emp = input("ID, Name, Position, Salary: ")
    f.write(emp + "\n")
    f.close()
    print("Employee added.\n")

def view_employees():
    try:
        f = open("employees.txt", "r")
        print("\nAll Employees:")
        print(f.read())
        f.close()
    except:
        print("No records found.\n")

def search_employee():
    emp_id = input("Enter Employee ID: ")
    found = False
    f = open("employees.txt", "r")
    for line in f:
        if line.startswith(emp_id + ","):
            print("Found:", line)
            found = True
    f.close()
    if not found:
        print("Not found.\n")

def update_employee():
    emp_id = input("Enter ID to update: ")
    lines = open("employees.txt", "r").readlines()
    f = open("employees.txt", "w")
    found = False
    for line in lines:
        if line.startswith(emp_id + ","):
            new = input("Enter new data (ID, Name, Position, Salary): ")
            f.write(new + "\n")
            found = True
        else:
            f.write(line)
    f.close()
    print("Updated.\n" if found else "Not found.\n")

def delete_employee():
    emp_id = input("Enter ID to delete: ")
    lines = open("employees.txt", "r").readlines()
    f = open("employees.txt", "w")
    found = False
    for line in lines:
        if not line.startswith(emp_id + ","):
            f.write(line)
        else:
            found = True
    f.close()
    print("Deleted.\n" if found else "Not found.\n")

while True:
    print("1. Add\n2. View\n3. Search\n4. Update\n5. Delete\n6. Exit")
    choice = input("Choose: ")
    if choice == "1": add_employee()
    elif choice == "2": view_employees()
    elif choice == "3": search_employee()
    elif choice == "4": update_employee()
    elif choice == "5": delete_employee()
    elif choice == "6": break
    else: print("Invalid choice.\n")
```

```python
def add():
    data = input("ID, Name, Position, Salary: ")
    with open("employees.txt", "a") as f:
        f.write(data + "\n")
    print("Added.\n")

def view():
    try:
        with open("employees.txt", "r") as f:
            for line in f:
                print(line.strip())
    except:
        print("No file.\n")

def search():
    id = input("Enter ID: ")
    found = False
    with open("employees.txt", "r") as f:
        for line in f:
            if line.startswith(id + ","):
                print("Found:", line.strip())
                found = True
    if not found:
        print("Not found.\n")

def update():
    id = input("ID to update: ")
    lines = open("employees.txt", "r").readlines()
    with open("employees.txt", "w") as f:
        for line in lines:
            if line.startswith(id + ","):
                new = input("New data (ID, Name, Position, Salary): ")
                f.write(new + "\n")
            else:
                f.write(line)
    print("Updated.\n")

def delete():
    id = input("ID to delete: ")
    lines = open("employees.txt", "r").readlines()
    with open("employees.txt", "w") as f:
        for line in lines:
            if not line.startswith(id + ","):
                f.write(line)
    print("Deleted.\n")

while True:
    print("1. Add\n2. View\n3. Search\n4. Update\n5. Delete\n6. Exit")
    c = input("Choose: ")
    if c == "1": add()
    elif c == "2": view()
    elif c == "3": search()
    elif c == "4": update()
    elif c == "5": delete()
    elif c == "6": break
    else: print("Invalid.\n")
```

Word Frequency Counter


```python
import os
import string
from collections import Counter

if not os.path.exists("sample.txt"):
    print("sample.txt not found.")
    print("Please enter a paragraph to create it:")
    text = input(">> ")
    with open("sample.txt", "w") as f:
        f.write(text)
else:
    with open("sample.txt", "r") as f:
        text = f.read()

text = text.lower()
for p in string.punctuation:
    text = text.replace(p, "")

words = text.split()
total_words = len(words)
counts = Counter(words)
top5 = counts.most_common(5)

print(f"\nTotal words: {total_words}")
print("Top 5 most common words:")
for word, count in top5:
    print(f"{word} - {count} time{'s' if count > 1 else ''}")

with open("word_count_report.txt", "w") as f:
    f.write("Word Count Report\n")
    f.write(f"Total Words: {total_words}\n")
    f.write("Top 5 Words:\n")
    for word, count in top5:
        f.write(f"{word} - {count}\n")
```

```python
import os
import string
from collections import Counter

def clean_text(line):
    line = line.lower()
    return line.translate(str.maketrans('', '', string.punctuation))
if not os.path.exists("sample.txt"):
    print("sample.txt topilmadi.")
    print("Iltimos, biror paragraf yozing:")
    with open("sample.txt", "w") as f:
        f.write(input(">> "))

word_counter = Counter()
total_words = 0

with open("sample.txt", "r") as f:
    for line in f:
        cleaned = clean_text(line)
        words = cleaned.split()
        word_counter.update(words)
        total_words += len(words)

try:
    top_n = int(input("Top nechta so‘zni ko‘rmoqchisiz? (masalan: 5): "))
except ValueError:
    top_n = 5  

top_common = word_counter.most_common(top_n)

print(f"\nTotal words: {total_words}")
print(f"Top {top_n} most common words:")
for word, count in top_common:
    print(f"{word} - {count} time{'s' if count > 1 else ''}")

with open("word_count_report.txt", "w") as f:
    f.write("Word Count Report\n")
    f.write(f"Total Words: {total_words}\n")
    f.write(f"Top {top_n} Words:\n")
    for word, count in top_common:
        f.write(f"{word} - {count}\n")
```
